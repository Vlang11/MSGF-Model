```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import seaborn as sns
import numpy as np
```

```python
xls = pd.ExcelFile(r'C:\Users\vlang\Desktop\Gust SampleSize\KSLE_SampleSize.xlsx')
```

```python
df1= pd.read_excel(xls, 'High Pressure')
df2 =pd.read_excel(xls, 'Convective')
df3 =pd.read_excel(xls, 'Night Overcast')
df4= pd.read_excel(xls, 'Non Convective')
df5 =pd.read_excel(xls, 'Less Windy Snow')
df6 =pd.read_excel(xls, 'Windy Snow')
df7= pd.read_excel(xls, 'Night Clear')
```

```python
nam24= df1[df1['N24bias'] !=999]
gfs24 = df1[df1['G24bias'] !=999]

sns.set_style(style='whitegrid')
fig, axes = plt.subplots(4, 2,   figsize=(20,16)) #Set up subplot axes
fig.suptitle('High Pressure for KSLE',fontsize=20)

sns.set(font_scale = 1.1)

sns.scatterplot(ax=axes[0, 0],x='ASOSspd', y='PPbias', data=df1).set(xlabel= 'Wind Speed (kt)', ylabel='bias (kt)')
axes[0,0].set_title("PProg",y=1.0, pad=-14)

sns.scatterplot(ax=axes[0, 1],x='ASOSdir', y='PPbias', data=df1).set(xlabel= 'Wind Direction (degree)', ylabel='bias (kt)')
axes[0,1].set_title("PProg",y=1.0, pad=-14)

sns.scatterplot(ax=axes[1, 0],x='ASOSspd', y='CLbia', data=df1).set(xlabel= 'Wind Speed (kt)', ylabel='bias (kt)')
axes[1,0].set_title("CLIM",y=1.0, pad=-14)

sns.scatterplot(ax=axes[1, 1],x='ASOSdir', y='CLbia', data=df1).set(xlabel= 'Wind Direction (degree)', ylabel='bias (kt)')
axes[1,1].set_title("CLIM",y=1.0, pad=-14)

sns.scatterplot(ax=axes[2, 0],x='ASOSspd', y='N24bias', data=nam24).set(xlabel= 'Wind Speed (kt)', ylabel='bias (kt)')
axes[2,0].set_title("NAM24",y=1.0, pad=-14)

sns.scatterplot(ax=axes[2, 1],x='ASOSdir', y='N24bias', data=nam24).set(xlabel= 'Wind Direction (degree)', ylabel='bias (kt)')
axes[2,1].set_title("NAM24",y=1.0, pad=-14)

sns.scatterplot(ax=axes[3, 0],x='ASOSspd', y='G24bias', data=nam24).set(xlabel= 'Wind Speed (kt)', ylabel='bias (kt)',ylim=(-30, 30))
axes[3,0].set_title("GFS24",y=1.0, pad=-14)

sns.scatterplot(ax=axes[3, 1],x='ASOSdir', y='G24bias', data=nam24).set(xlabel= 'Wind Direction (degree)', ylabel='bias (kt)',ylim=(-30, 30))
axes[3,1].set_title("GFS24",y=1.0, pad=-14)

```
